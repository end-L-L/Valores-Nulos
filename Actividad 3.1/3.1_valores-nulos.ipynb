# Sustitución de Valores Nulos

```python
import pandas as pd
```

```python
df = pd.read_csv('VentasTotales.csv')
df.head()
```

```python
df.info()
```

```python
# Identificar valores nulos por columna
valores_nulos = df.isnull().sum()
valores_nulos.to_string("valores_nulos.txt")
valores_nulos
```

```python
# Identificar valores nulos por data frame
valores_nulos = df.isnull().sum().sum()
valores_nulos
```

```python
df1 = df.copy()
df1.head()
```

# Sustitución

```python
# Filtro 1
# Sustituir los valores nulos de la columna 'salon_ventas' con promedio (media)
df1['salon_ventas'] = df1['salon_ventas'].fillna(round(df['salon_ventas'].mean(),1))
df1.to_csv('filtro1.csv')
df1
```

```python
# Filtro 2
# Sustituir los valores nulos de la columna 'tarjetas_debito' con la mediana
df1['tarjetas_debito'] = df1['tarjetas_debito'].fillna(round(df['tarjetas_debito'].median(),1))
df1.to_csv('filtro2.csv')
df1
```

```python
# Filtro 3
# Sustituir los valores nulos de la columna 'tarjetas_credito' con el promedio (media)
df1['tarjetas_credito'] = df1['tarjetas_credito'].fillna(round(df['tarjetas_credito'].mean(),1))
df1.to_csv('filtro3.csv')
df1
```

```python
# Filtro 4
# Sustituir los valores nulos de la columna 'otros_medios' con la mediana
df1['otros_medios'] = df1['otros_medios'].fillna(round(df['otros_medios'].median(),1))
df1.to_csv('filtro4.csv')
df1
```

```python
# Filtro 5
# Sustituir los valores nulos de la columna 'subtotal_ventas_alimentos_bebidas' con el promedio (media)
df1['subtotal_ventas_alimentos_bebidas'] = df1['subtotal_ventas_alimentos_bebidas'].fillna(101.0)
df1.to_csv('filtro5.csv')
df1
```

```python
# Filtro 6
# Sustituir los valores nulos de la columna 'bebidas' por valores no nulos hacia adelante "forward fill"
df1['bebidas'] = df1['bebidas'].ffill()
df1.to_csv('filtro6.csv')
df1
```

```python
# Filtro 7
# Sustituir los valores nulos de la columna 'almacen' por valores no nulos hacia atras "backward fill"
df1['almacen'] = df1['almacen'].bfill()
df1.to_csv('filtro7.csv')
df1
```

```python
# Filtro 8
# Sustituir los valores nulos de la columna 'panaderia' por valores no nulos hacia adelante "forward fill"
df1['panaderia'] = df1['panaderia'].ffill()
df1.to_csv('filtro8.csv')
df1
```

```python
# Filtro 9
# Sustituir los valores nulos de la columna 'lacteos' por valores no nulos hacia atras "backward fill"
df1['lacteos'] = df1['lacteos'].bfill()
df1.to_csv('filtro9.csv')
df1
```

```python
# Filtro 10
# Sustituir los valores nulos de la columna 'carnes' por valores no nulos hacia adelante "forward fill"
df1['carnes'] = df1['carnes'].ffill()
df1.to_csv('filtro10.csv')
df1
```

```python
# Filtro 11
# Sustituir los valores nulos de la columna 'verduleria_fruteria' por valores no nulos hacia atras "backward fill"
df1['verduleria_fruteria'] = df1['verduleria_fruteria'].bfill()
df1.to_csv('filtro11.csv')
df1
```

```python
# Filtro 12
# Sustituir los valores nulos de la columna 'alimentos_preparados_rotiseria' por valores no nulos hacia adelante "forward fill"
df1['alimentos_preparados_rotiseria'] = df1['alimentos_preparados_rotiseria'].ffill()
df1.to_csv('filtro12.csv')
df1
```

```python
# Filtro 13
# Sustituir los valores nulos de la columna 'indumentaria_calzado_textiles_hogar' con el promedio (media)
df1['indumentaria_calzado_textiles_hogar'] = df1['indumentaria_calzado_textiles_hogar'].fillna(round(df['indumentaria_calzado_textiles_hogar'].mean(),1))
df1.to_csv('filtro13.csv')
df1
```

```python
# Filtro 14
# Sustituir los valores nulos de la columna 'electronicos_articulos_hogar' con la mediana
df1['electronicos_articulos_hogar'] = df1['electronicos_articulos_hogar'].fillna(round(df['electronicos_articulos_hogar'].median(),1))
df1.to_csv('filtro14.csv')
df1
```

```python
# Filtro 15
# Sustituir los valores nulos de la columna 'otros' por un valor numerico en concreto
df1['otros'] = df1['otros'].fillna(0)
df1.to_csv('filtro15.csv')
df1
```

```python
valores_nulos = df1.isnull().sum()
valores_nulos.to_string("valores_nulos_final.txt")
valores_nulos
```
